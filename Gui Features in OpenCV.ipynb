# Gui Features in OpenCV

### Read an image

```python
import numpy as np
import cv2
"""
Mode:
    cv2.IMREAD_COLOR: 1, default
    cv2.IMREAD_GRAYSCALE: 0
    cv2.IMREAD_UNCHANGED: -1 (Loads image as such including alpha channel)
"""
img = cv2.imread('./images/squirrel.jpg', 0)
```

### Display an image

```python
# default is `cv2.WINDOW_AUTOSIZE`
cv2.namedWindow('squirrel', cv2.WINDOW_NORMAL)
cv2.imshow('squirrel', img)
cv2.waitKey(2000) # wait for a key to exit.
# cv2.destroyAllWindows()
cv2.destroyWindow('squirrel')
```

### Write an image

```python
import numpy as np
import cv2

img = cv2.imread('./images/squirrel.jpg', 0)
cv2.imshow('squirrel', img)
k = cv2.waitKey(0) & 0xff # on 65-bit sys.
if k == 27: # wait for ESC key to exit
    cv2.destroyAllWindows()
elif k == ord('s'):
    cv2.imwrite('./images/squirrel2.png', img)
    cv2.destroyAllWindows()
```

### Using Matplotlib

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('./images/squirrel.jpg', 0)
plt.imshow(img, cmap='gray', interpolation='bicubic')
plt.xticks([]), plt.yticks([]) # to hide tick values on X and Y axis
plt.show()
```
